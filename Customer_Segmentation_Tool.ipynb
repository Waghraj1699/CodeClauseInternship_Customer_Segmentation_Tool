```python
new_df = dataset.sample(300000,random_state=42)
```

```python
new_df.dropna(inplace=True)
```

```python
# creating mrf table

new_df['InvoiceDate'] = pd.to_datetime(new_df['InvoiceDate'])
```

```python
new_df.head()
```

```python
 maxDate = max(new_df['InvoiceDate'])
```

```python
maxDate
```

```python
new_df['diff'] = maxDate - new_df['InvoiceDate']
```

```python
new_df.head()
```

```python
# calculate Recency

rfm_r = new_df.groupby('CustomerID')['diff'].min()
rfm_r = rfm_r.reset_index()
```

```python
rfm_r.head()
```

```python
rfm_r['diff'] = rfm_r['diff'].dt.days
```

```python
rfm_r.head()
```

```python
# calculate MonetaryValue	

new_df['TotalPrice'] = new_df['Quantity'] * new_df['UnitPrice']
rfm_m = new_df.groupby('CustomerID')['TotalPrice'].sum()
rfm_m = rfm_m.reset_index()
```

```python
rfm_m.head()
```

```python
rfm = pd.merge(rfm_r,rfm_m,on='CustomerID')
rfm.head()
```

```python
new_df.head()
```

```python
# calculate Frequency

rfm_f = new_df.groupby('CustomerID')['InvoiceNo'].count()
rfm_f = rfm_f.reset_index()
rfm_f.InvoiceNo.values
```

```python
new_rfm = pd.merge(rfm,rfm_f,on='CustomerID')
new_rfm.head()
```

```python
new_rfm.rename(columns = {'diff': 'Recency','InvoiceNo': 'Frequency',
                            'TotalPrice': 'MonetaryValue'}, inplace=True)
```

```python
new_rfm.head()
```

```python
# outlier analysis

col = ['Recency','MonetaryValue','Frequency']

plt.rcParams['figure.figsize'] = [15,6]
sns.boxplot(data=new_rfm[col],orient='v',palette='Set2',whis=1.5,saturation=1,width=0.7)
plt.title('Outliers Distribution',fontsize=14,fontweight='bold')
plt.xlabel('Range',fontweight='bold')
plt.ylabel('columns',fontweight='bold')
plt.show()
```

```python
# statistical analysis
# removing outliers
# Q - > Quantil
# IQR -> Inter quantile range
```

```python
Q1 = new_rfm.MonetaryValue.quantile(0.05)
Q3 = new_rfm.MonetaryValue.quantile(0.95)
IQR = Q3 - Q1
new_rfm = new_rfm[(new_rfm.MonetaryValue >= Q1 - 1.5*IQR ) & (new_rfm.MonetaryValue <= Q3 + 1.5*IQR )]
```

```python
Q1 = new_rfm.Recency.quantile(0.05)
Q3 = new_rfm.Recency.quantile(0.95)
IQR = Q3 - Q1
new_rfm = new_rfm[(new_rfm.Recency >= Q1 - 1.5*IQR ) & (new_rfm.Recency <= Q3 + 1.5*IQR )]
```

```python
Q1 = new_rfm.Frequency.quantile(0.05)
Q3 = new_rfm.Frequency.quantile(0.95)
IQR = Q3 - Q1
new_rfm = new_rfm[(new_rfm.Frequency >= Q1 - 1.5*IQR ) & (new_rfm.Frequency <= Q3 + 1.5*IQR )]
```

```python
new_rfm.head()
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
rfm_df = new_rfm[['Recency','MonetaryValue','Frequency']]

scaled_rfm = sc.fit_transform(rfm_df)
```

```python
scaled_rfm
```

```python
scaled_rfm = pd.DataFrame(scaled_rfm)
```

```python
scaled_rfm
```

```python
scaled_rfm.columns = ['Recency','MonetaryValue','Frequency']
```

```python
scaled_rfm
```

```python
# training the k-mean algorithm

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4,max_iter=40)
kmeans.fit(scaled_rfm)
```

```python
kmeans.labels_
```

```python
set(kmeans.labels_)
```

```python
# ELBOW method to find the right number of clusters
```

```python
inertia = []

for num_clust in range(1,10):
    kmeans = KMeans(n_clusters=num_clust,max_iter=40)
    kmeans.fit(scaled_rfm)

    inertia.append(kmeans.inertia_)

plt.plot(inertia)

    
```

```python
# Training Kmeans with 3 cluster
```

```python
kmeans = KMeans(n_clusters=3,max_iter=500)
kmeans.fit(scaled_rfm)
```
